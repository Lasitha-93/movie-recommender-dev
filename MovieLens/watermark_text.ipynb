
# Text watermark


Use a Text as a watermark



```python
import numpy as np
#import matplotlib
#matplotlib.use('Agg')
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

# Fixing random state for reproducibility
np.random.seed(19680801)


fig,ax = plt.subplots()
ax.plot(['a','b','c','d','e'],[5,3,8,6,1], '-o', ms=5, lw=2, color='green', alpha=0.7, mfc='orange')
ax.plot(['a','b','c','d','f'],[45,53,68,76,31], '-o', ms=5, color='red',lw=2, alpha=0.7, mfc='orange')
ax.grid()

red_patch = mpatches.Patch(color='red', label='The red data')
green_patch = mpatches.Patch(color='green', label='The green data')
plt.legend(handles=[red_patch,green_patch])

# position bottom right
# fig.text(0.95, 0.02, 'Lasitha Randunu',
#          fontsize=50, color='gray',
#          ha='right', va='bottom', alpha=1)

plt.show()
```
